```python
import joblib
import numpy as np
import pandas as pd

model = joblib.load("../models/higgs_xgboost.pkl")
```

```python
# Column names as defined in the Higgs dataset
low_feature = ['lepton_pT', 'lepton_eta', 'lepton_phi', 
        'missing_energy_magnitude', 'missing_energy_phi',
        'jet1_pt', 'jet1_eta', 'jet1_phi', 'jet1_b-tag',
        'jet2_pt', 'jet2_eta', 'jet2_phi', 'jet2_b-tag',
        'jet3_pt', 'jet3_eta', 'jet3_phi', 'jet3_b-tag',
        'jet4_pt', 'jet4_eta', 'jet4_phi', 'jet4_b-tag']

high_feature = ['m_jj', 'm_jjj', 'm_lv', 'm_jlv', 'm_bb', 'm_wbb', 'm_wwbb']

columns = ['label'] + low_feature + high_feature

# Load only first 100,000 rows to keep it fast


df_test = pd.read_csv(
    "../data/HIGGS.csv.gz",
    compression="gzip",
    nrows=10,
    names = columns,
    header=None
)

X_new = df_test.iloc[:, 1:]
```

```python
pred_class = model.predict(X_new)
pred_prob = model.predict_proba(X_new)[:, 1]

print("Predicted class:", pred_class)
print("Signal probability:", pred_prob)
```
